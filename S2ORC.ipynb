```python
# a function to read a discpline into a list

def doctolist(subject):
    BATCH_ID = [0, 1 ,2]



    texts = []

    for id in BATCH_ID:
        with open(f'pdf_parses/{subject}/pdf_parses_{id}.jsonl.gz') as f_pdf:
            for line in f_pdf:
                pdf_parse_dict = json.loads(line)        
                text = [i['text'] for i in pdf_parse_dict['body_text']]
                texts += text
    return texts
    
```

```python
import nltk
import itertools

def listtotokens(tokens):
    # nltk.regexp_tokenize() allows you to build a customized tokenizer by specifying a regex pattern
    pattern = r'''(?x)    # set flag to allow verbose regexps
    \w+
    (?:[A-Z]\.)+          # abbreviations, e.g. U.S.A.
    |\w+(?:-\w+)*         # words with optional internal hyphens
    |\$?\d+(?:\.\d+)?%?   # currency and percentages, e.g. $12.40, 82%
    |\.\.\.               # ellipsis
    |[][.,;"\'?():-_`]    # these are separate tokens
    '''
    
    
    token = []
    for t in subject:
        token.append(nltk.regexp_tokenize(t, pattern))
    
    tokens = list(itertools.chain.from_iterable(token))

    filtered_tokens = []
    punc = ["'",".", ":", ";",",", "?","(",")","[","]","\""]
    for token in tokens:
        if token not in punc:
            filtered_tokens.append(token)

    f=open(f'{subject}-tokens.txt', 'w')

    for ele in filtered_tokens:
        f.write(ele + '\n')
    
    f.close()
    
    
    
    
    
    
```

```python
def preprocess(subject):
    texts = doctolist(subject)
    listtotokens(texts)
```
